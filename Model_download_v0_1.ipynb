```python
# Download the model to Colab
from huggingface_hub import snapshot_download

# Example model (replace with your desired model)
model_id =   "meta-llama/Llama-2-7b-chat-hf" #"meta-llama/Meta-Llama-3-8B" #"orel12/ggml-gpt4all-j-v1.3-groovy"  #"mistralai/Mistral-7B-Instruct-v0.3"

# Download model files
model_path = snapshot_download(
    repo_id=model_id,
    local_dir="/content/downloaded_model"
)

print(f"Model downloaded to: {model_path}")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained("/content/downloaded_model")
# Add a padding token if it doesn't exist
if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': '[PAD]'})
# Load the model
model = AutoModelForCausalLM.from_pretrained("/content/downloaded_model", torch_dtype="auto")
# Resize the model embeddings to account for the new pad token if added
model.resize_token_embeddings(len(tokenizer))
```

```python
llm = model
inputs = tokenizer("Explain how neural networks work in simple terms", return_tensors="pt")
outputs = llm.generate(**inputs, max_length=50)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

```python

# Print model size
model_size = sum(p.numel() for p in llm.parameters())
print(f"Model size: {model_size} parameters")

# Print model size in MB
model_size_in_mb = model_size * 4 / (1024 ** 2)  # assuming 4 bytes per parameter
print(f"Model size: {model_size_in_mb:.2f} MB")

# Print model size in GB
model_size_in_gb = model_size * 4 / (1024 ** 3)  # assuming 4 bytes per parameter
print(f"Model size: {model_size_in_gb:.2f} GB")

# #"mistralai/Mistral-7B-Instruct-v0.3"
# Model size: 7248031744 parameters
# Model size: 27649.05 MB
# Model size: 27.00 GB

# meta-llama/Meta-Llama-3-8B
# Model size: 8030269440 parameters
# Model size: 30633.05 MB
# Model size: 29.92 GB
```
